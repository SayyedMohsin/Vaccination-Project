```python
# üìå Test Cleaned Vaccination Data
import pandas as pd
import matplotlib.pyplot as plt
import os

EXPORT = "exports"

# Load all tables
df_country   = pd.read_csv(os.path.join(EXPORT, "dim_country.csv"))
df_schedule  = pd.read_csv(os.path.join(EXPORT, "dim_schedule.csv"))
df_intro     = pd.read_csv(os.path.join(EXPORT, "dim_vaccine_intro.csv"))
df_cases     = pd.read_csv(os.path.join(EXPORT, "fact_cases.csv"))
df_incidence = pd.read_csv(os.path.join(EXPORT, "fact_incidence.csv"))
df_cov       = pd.read_csv(os.path.join(EXPORT, "fact_coverage.csv"))

print("‚úÖ All cleaned data loaded")

# ------------------------
# Q1: Coverage vs Year
# ------------------------
q1 = df_cov.groupby("year")["coverage_percent"].mean().reset_index()
display(q1.head(10))

plt.figure(figsize=(8,4))
plt.plot(q1["year"], q1["coverage_percent"], marker="o")
plt.title("Average Coverage vs Year")
plt.xlabel("Year")
plt.ylabel("Coverage (%)")
plt.show()

# ------------------------
# Q2: Drop-off Dose1 ‚Üí Dose2
# ------------------------
q2 = df_cov.pivot_table(index=["antigen","year"],
                        columns="coverage_category",
                        values="coverage_percent",
                        aggfunc="mean").reset_index()

if "DOSE1" in q2 and "DOSE2" in q2:
    q2["drop_off"] = q2["DOSE1"] - q2["DOSE2"]
    display(q2.head(10))

    plt.figure(figsize=(8,4))
    plt.bar(q2["year"], q2["drop_off"])
    plt.title("Drop-off Rate (Dose1 ‚Üí Dose2)")
    plt.xlabel("Year")
    plt.ylabel("Drop-off (%)")
    plt.show()
else:
    print("‚ö†Ô∏è DOSE1/DOSE2 columns not found in coverage data")

# ------------------------
# Q3: Cases vs Incidence Correlation
# ------------------------
q3 = df_cases.groupby("year")["cases"].sum().reset_index()
q3_inc = df_incidence.groupby("year")["incidence_rate"].mean().reset_index()
q3 = q3.merge(q3_inc, on="year", how="inner")

display(q3.head(10))

plt.figure(figsize=(8,4))
plt.scatter(q3["cases"], q3["incidence_rate"])
plt.title("Reported Cases vs Incidence Rate")
plt.xlabel("Cases")
plt.ylabel("Incidence Rate")
plt.show()
```

```python
import pandas as pd
import os

# Absolute path set ‡§ï‡§∞‡•ã
EXPORT = r"C:\JN\Vaccination-Project\exports"

# ‡§Ö‡§¨ ‡§∏‡§≠‡•Ä CSV load ‡§ï‡§∞‡•á‡§Ç
df_country   = pd.read_csv(os.path.join(EXPORT, "dim_country.csv"))
df_schedule  = pd.read_csv(os.path.join(EXPORT, "dim_schedule.csv"))
df_intro     = pd.read_csv(os.path.join(EXPORT, "dim_vaccine_intro.csv"))
df_cases     = pd.read_csv(os.path.join(EXPORT, "fact_cases.csv"))
df_incidence = pd.read_csv(os.path.join(EXPORT, "fact_incidence.csv"))
df_coverage  = pd.read_csv(os.path.join(EXPORT, "fact_coverage.csv"))

# ‚úÖ Quick check - ‡§π‡§∞ table ‡§Æ‡•á‡§Ç ‡§ï‡§ø‡§§‡§®‡•Ä rows ‡§π‡•à‡§Ç
print("dim_country:", len(df_country))
print("dim_schedule:", len(df_schedule))
print("dim_vaccine_intro:", len(df_intro))
print("fact_cases:", len(df_cases))
print("fact_incidence:", len(df_incidence))
print("fact_coverage:", len(df_coverage))

# ‚úÖ First rows ‡§¶‡§ø‡§ñ‡§æ‡§®‡•á ‡§ï‡•á ‡§≤‡§ø‡§è
print("\n--- dim_country ---")
print(df_country.head())

print("\n--- fact_coverage ---")
print(df_coverage.head())
```

```python
# ---- dim_country ----
df_country = df_country.dropna(subset=["iso3", "country_name"])
df_country["who_region"] = df_country["who_region"].fillna("Unknown")

# ---- fact_coverage ----
df_coverage["year"] = df_coverage["year"].astype("Int64")  # float -> int
df_coverage = df_coverage.dropna(subset=["coverage_percent"])
df_coverage["coverage_category_desc"] = df_coverage["coverage_category_desc"].fillna("Unknown")

# ---- dim_schedule ----
df_schedule["schedule_round"] = df_schedule["schedule_round"].fillna("Unknown")
df_schedule["age_admin"] = df_schedule["age_admin"].fillna("Unknown")

# ---- dim_vaccine_intro ----
df_intro["intro_year"] = pd.to_numeric(df_intro["intro_year"], errors="coerce")
df_intro = df_intro.dropna(subset=["intro_year"])
df_intro["intro_year"] = df_intro["intro_year"].astype(int)

# ---- fact_cases ----
df_cases["year"] = df_cases["year"].astype("Int64")
df_cases["disease_desc"] = df_cases["disease_desc"].fillna("Unknown")

# ---- fact_incidence ----
df_incidence["year"] = df_incidence["year"].astype("Int64")
df_incidence["disease_desc"] = df_incidence["disease_desc"].fillna("Unknown")

# ‚úÖ Check summary after cleaning
print("dim_country:", df_country.shape)
print("fact_coverage:", df_coverage.shape)
print("dim_schedule:", df_schedule.shape)
print("dim_vaccine_intro:", df_intro.shape)
print("fact_cases:", df_cases.shape)
print("fact_incidence:", df_incidence.shape)
```

```python
# Cleaned export folder ‡§¨‡§®‡§æ‡§ì
EXPORT_CLEAN = "../exports_clean"
import os
os.makedirs(EXPORT_CLEAN, exist_ok=True)

# Export ‡§π‡§∞ table ‡§ï‡•ã CSV ‡§Æ‡•á‡§Ç
df_country.to_csv(os.path.join(EXPORT_CLEAN, "dim_country_clean.csv"), index=False)
df_schedule.to_csv(os.path.join(EXPORT_CLEAN, "dim_schedule_clean.csv"), index=False)
df_intro.to_csv(os.path.join(EXPORT_CLEAN, "dim_vaccine_intro_clean.csv"), index=False)
df_cases.to_csv(os.path.join(EXPORT_CLEAN, "fact_cases_clean.csv"), index=False)
df_incidence.to_csv(os.path.join(EXPORT_CLEAN, "fact_incidence_clean.csv"), index=False)
df_coverage.to_csv(os.path.join(EXPORT_CLEAN, "fact_coverage_clean.csv"), index=False)

print("‚úÖ All cleaned CSVs exported to:", EXPORT_CLEAN)
```
